```python
# Week 07 Experiment: Simple Local Port Scanner (SAFE)
import socket

def simple_local_scan(ports):
    host = "127.0.0.1"  # localhost only (safe)
    print(f"Scanning {host}...\n")
    
    for port in ports:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.settimeout(0.5)
        
        result = sock.connect_ex((host, port))
        
        if result == 0:
            print(f"Port {port} is OPEN")
        else:
            print(f"Port {port} is closed")
        
        sock.close()

# Common local ports to test
ports_to_test = [21, 22, 80, 443, 3306]

simple_local_scan(ports_to_test)
```
