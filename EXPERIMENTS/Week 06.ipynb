```python
# Experiment 1: Hash comparison + avalanche effect
import hashlib

sample = r"C:\Users\aisha\Documents\UNIVERSITY\E-Portfolio- Networks and System Security\EXPERIMENTS\test.txt"

def compute(path, algo):
    h = hashlib.new(algo)
    with open(path, "rb") as f:
        h.update(f.read())
    return h.hexdigest()

print("MD5:   ", compute(sample, "md5"))
print("SHA1:  ", compute(sample, "sha1"))
print("SHA256:", compute(sample, "sha256"))
```

```python
# Experiment 2: Extract first 20 strings from a PE file
import re

sample = r"C:\Users\aisha\Documents\UNIVERSITY\E-Portfolio- Networks and System Security\EXPERIMENTS\test.txt"

def get_strings(path):
    with open(path, "rb") as f:
        data = f.read()
    return re.findall(rb"[ -~]{4,}", data)

strings = get_strings(sample)

for s in strings[:20]:
    print(s.decode(errors="ignore"))
```

```python
# Experiment 4: Simple YARA rule
import yara

sample = r"C:\Users\aisha\Documents\UNIVERSITY\E-Portfolio- Networks and System Security\EXPERIMENTS\test.txt"

rule = yara.compile(source="""
rule ContainsHTTP {
    strings:
        $s = "http"
    condition:
        $s
}
""")

print(rule.match(sample))
```
