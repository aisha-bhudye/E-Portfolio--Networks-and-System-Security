```python
# Week 03 Experiment B: TOTP drift tolerance
import pyotp
import time

totp = pyotp.TOTP(pyotp.random_base32())
print("Secret:", totp.secret)

print("\nCodes at different time offsets:")
for drift in [-30, -15, 0, 15, 30]:
    print(f"Time drift {drift:+} sec:", totp.at(int(time.time()) + drift))
```

```python
# Week 03 Experiment A: bcrypt cost vs time
import bcrypt
import time

password = b"password123"
costs = [8, 10, 12, 14]

for c in costs:
    print(f"\n=== Cost {c} ===")
    t1 = time.time()
    bcrypt.hashpw(password, bcrypt.gensalt(rounds=c))
    t2 = time.time()
    print("Hash time:", round((t2 - t1) * 1000, 3), "ms")
```
