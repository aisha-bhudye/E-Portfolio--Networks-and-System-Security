```python
# Week 02 Experiment: RSA timing test 
import time
from Crypto.PublicKey import RSA
from Crypto.Cipher import PKCS1_OAEP
import numpy as np

key_sizes = [1024, 2048, 3072, 4096]
message = b"Hello World!"

for size in key_sizes:
    print(f"\n=== Testing RSA-{size} ===")
    key = RSA.generate(size)
    cipher = PKCS1_OAEP.new(key.publickey())
    decipher = PKCS1_OAEP.new(key)

    # Encrypt timing
    t1 = time.time()
    for _ in range(200):
        ct = cipher.encrypt(message)
    t2 = time.time()

    # Decrypt timing
    t3 = time.time()
    for _ in range(200):
        decipher.decrypt(ct)
    t4 = time.time()

    print("Encrypt avg ms:", round((t2 - t1) / 200 * 1000, 4))
    print("Decrypt avg ms:", round((t4 - t3) / 200 * 1000, 4))
```
