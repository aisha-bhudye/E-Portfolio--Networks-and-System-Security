```python
# Week 04 Experiment: File integrity checker (safe)
import hashlib
import os
import time

FILE = "testfile.txt"

# Create file if not exists
if not os.path.exists(FILE):
    with open(FILE, "w") as f:
        f.write("This is a test file.\n")

def sha256(path):
    with open(path, "rb") as f:
        return hashlib.sha256(f.read()).hexdigest()

baseline = sha256(FILE)
print("Baseline hash:", baseline)
print("Modify the file manually to test detection...\n")

while True:
    time.sleep(2)
    new_hash = sha256(FILE)
    if new_hash != baseline:
        print(" File changed! New hash:", new_hash)
        break
    else:
        print("No change detected...")
```
