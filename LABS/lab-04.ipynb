```python

import hashlib
import os
import csv
from datetime import datetime

folder_path = "sample_folder"
output_file = "file_hashes.csv"

with open(output_file, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["File Name", "SHA256 Hash", "Timestamp"])
    
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            with open(file_path, "rb") as f:
                file_hash = hashlib.sha256(f.read()).hexdigest()
            writer.writerow([filename, file_hash, datetime.now().isoformat()])

print(f"Hashes saved to {output_file}")
```

```python


import csv
import hashlib
import os

folder_path = "sample_folder"
baseline_file = "file_hashes.csv"

# Load baseline hashes into a dictionary
baseline_hashes = {}
with open(baseline_file, "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        baseline_hashes[row["File Name"]] = row["SHA256 Hash"]

changes = []
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        with open(file_path, "rb") as f:
            current_hash = hashlib.sha256(f.read()).hexdigest()
        if filename not in baseline_hashes or current_hash != baseline_hashes[filename]:
            changes.append(filename)

print("Changed or new files:", changes)
```

```python

import re
import os

SIGNATURES = [r"eval\\(", r"base64\\.b64decode", r"socket\\.connect", r"exec\\(", r"import os"]
folder_path = "sample_folder"

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        with open(file_path, "r", errors="ignore") as f:
            content = f.read()
            for sig in SIGNATURES:
                if re.search(sig, content):
                    print(f"Suspicious pattern '{sig}' found in {filename}")
```

```python

import pandas as pd
import hashlib
import os

folder_path = "sample_folder"
baseline_file = "file_hashes.csv"

baseline = pd.read_csv(baseline_file)
changes = []

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        with open(file_path, "rb") as f:
            current_hash = hashlib.sha256(f.read()).hexdigest()
        old_hash = baseline.loc[baseline["File Name"] == filename, "SHA256 Hash"].values
        if len(old_hash) == 0 or current_hash != old_hash[0]:
            changes.append(filename)

print("Changed or new files:", changes)
```
